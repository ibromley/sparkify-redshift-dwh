```python
%%sql
    INSERT INTO songplays (
        ts, 
        user_id, 
        level, 
        song_id, 
        artist_id, 
        session_id, 
        location, 
        user_agent
    )
    SELECT DISTINCT timestamp 'epoch' + se.ts/1000 * interval '1 second' AS ts, 
		se.userid, 
        se.level, 
        s.song_id, 
        a.artist_id, 
        se.sessionid, 
        se.location, 
        se.useragent
    FROM staging_events se
     JOIN songs s ON (s.title = se.song)
     JOIN artists a ON (a.artist_name = se.artist_name)
    WHERE s.artist_id = a.artist_id
```

```python
%%sql
SELECT * FROM songplays
```
